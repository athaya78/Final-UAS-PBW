```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime

from google.colab import files
uploaded = files.upload()

```

```python
dataset = pd.read_csv('supermarket_sales - Sheet1.csv')
dataset['order_month'] = dataset['Date'].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y").strftime('%Y-%m'))
#atur ukuran grafik
fig = plt.figure(figsize=(15,5))
#visualisasi
dataset.groupby('order_month')['Total'].sum().plot(color='green', marker='o', linestyle='-.', linewidth=2)
plt.title('Sales Report in Quarter I', loc='center', pad=20, color='blue', fontsize=24)
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total sales', fontsize=16)
plt.grid(color='darkgray', linestyle=':', linewidth='1')
plt.text(0.4, 0.2 , 'The sales decreased significantly on February 2019', transform=fig.transFigure, color='red', fontsize=12)
print(dataset.head(10))
```

```python
#visualisasi
plt.figure(figsize=(15,5))
dataset[dataset['order_month']=='2019-02'].groupby(['Date'])['Invoice ID'].nunique().plot(color='red', marker='.', linewidth=2)
plt.title('Daily Number of Customers - February 2019', loc='center', pad=20, color='orange', fontsize=24)
plt.xlabel('Order Date', fontsize=15, color='blue')
plt.ylabel('Number of Customers', fontsize=15, color='blue')
plt.grid(color='darkgray', linestyle=':', linewidth='0.5')
plt.ylim(ymin=0)
plt.show()
```
